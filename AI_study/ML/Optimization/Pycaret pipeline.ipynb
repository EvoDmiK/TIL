```python
from pycaret.datasets import get_data
from pycaret.classification import *
```

```python
dataset = get_data('credit')
train   = dataset.sample(frac = 0.95, random_state = 42)
test    = dataset.drop(train.index)

train.reset_index(inplace = True, drop = True)
test.reset_index(inplace  = True, drop = True)
```

```python
exp_clf = setup(data = train, target = 'default', session_id = 99)
```

```python
best_model = compare_models()
```

```python
rf = create_model('rf')
```

```python
tuned_rf = tune_model(rf)
```

```python
dt = create_model('dt')
rf = create_model('rf')

blender      = blend_models(estimator_list = [dt, rf])
best_model_5 = compare_models(n_select = 5)
blender5     = blend_models(best_model_5)
```

```python
final_model = finalize_model(blender5)
prediction  = predict_model(final_model, data = test)
```
