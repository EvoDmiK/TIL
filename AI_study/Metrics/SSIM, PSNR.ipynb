```python
import os

from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from imutils.paths import list_images
from matplotlib import pyplot as plt
from skimage.transform import resize
import imquality.brisque as brisque
from skimage.io import imread
import numpy as np
```

```python
SEP       = os.path.sep
ROOT_PATH = SEP.join(os.getcwd().split(SEP)[:-2])
DATA_PATH = f'{ROOT_PATH}/Datasets/phenaki'
```

```python
def show_image(image):

    plt.imshow(image)
    plt.axis(False)
```

```python
image_paths = sorted(list_images(DATA_PATH))
sample      = imread(image_paths[-1])
```

```python
show_image(sample)
```

```python
for image_path in image_paths:
    file_name    = image_path.split(SEP)[-1]
    file_name, _ = os.path.splitext(file_name)
    
    image  = imread(image_path)
    origs  = [image[:131, idx * 131 : (idx + 1) * 131]
              for idx in range(11)]
    recons = [image[131 : 262, idx * 131 : (idx + 1) * 131]
              for idx in range(11)] 

    SSIM, PSNR, BRISQUE = [], [], []
    for ori, recon in zip(origs, recons):

        # ori   = resize(ori, (256, 256))
        # recon = resize(recon, (256, 256))

        print(ori.shape, recon.s)
        PSNR.append(psnr(ori, recon))
        SSIM.append(ssim(ori, recon, channel_axis = 2))


    print(f'[{file_name}] PSNR : {np.mean(PSNR) :.3f}')
    print(f'[{file_name}] SSIM : {np.mean(SSIM) :.3f}')
    print('\n\n')
```
