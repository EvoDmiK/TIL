```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
IMG_SIZE  = (200, 200)

def convolution(image, pool_size = 5):
    
    H, W = image.shape
    filter = np.random.normal(size = (pool_size, pool_size))
    
    for h_idx in range(H - (pool_size - 1)):
        for w_idx in range(W - (pool_size -1)):
            partial = image[w_idx: w_idx + pool_size, h_idx:h_idx + pool_size]
            filtered = np.dot(partial, filter)
            
            image[w_idx: w_idx + pool_size, h_idx: h_idx + pool_size] = filtered
    return image
```

```python
image = np.ones(IMG_SIZE, dtype = np.int8)

image = np.ones(IMG_SIZE, dtype = np.int8)
print(image)

plt.imshow(image, cmap = 'gray')
plt.show()

image = convolution(image*255, pool_size = 10)

plt.imshow(image, cmap = 'gray')
plt.show()            

cv2.write('dd.png', image)
```
