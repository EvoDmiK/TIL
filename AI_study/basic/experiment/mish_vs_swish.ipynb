```python
import matplotlib.pyplot as plt
from sympy import *
import numpy as np

sym_x, sym_y = symbols('x y')
init_printing()

x = np.arange(-10.0, 10.0, 0.01)

mish_form = sym_x*(2+exp(sym_x))/(2+exp(sym_x)*(2+exp(sym_x)))
```

```python
%time
mish_lamb = lambda x: x*((np.exp(x)*(2+np.exp(x)))/ (2 + np.exp(x)*(2+np.exp(x))))
output_mish = mish_lamb(x)

mish_form
```

```python
%time
swish = lambda x: x/(1+np.exp(-x))
output_swish = swish(x)
```

```python
%time
plt.plot(x, output_mish)
plt.ylim(-1, 2.5)
plt.axvline(0.0, color='gray', linestyle='--', alpha = 0.8)
plt.axhline(0.0, color='gray', linestyle='--', alpha = 0.8)
plt.show()

mish_form
```
