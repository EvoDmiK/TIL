```python
from collections import Counter
```

```python
def solution(scores):
    
    answer = ''
    for my, _ in enumerate(scores):
        my_score = [scores[oth][my] for oth, _ in enumerate(scores)]
        score_count = Counter(my_score)
        
        my_score_count = score_count[scores[my][my]]
        self_score = scores[my][my]
        
        condition = [my_score_count == 1, any([self_score == min(my_score), self_score == max(my_score)])]
                        
        if all(condition): my_score.remove(self_score)
            
        avg = sum(my_score) / len(my_score)
        print(my_score, avg)

        if avg >= 90: grade = 'A'
        elif 90 > avg >= 80: grade = 'B'
        elif 80 > avg >= 70: grade = 'C'
        elif 70 > avg >= 50: grade = 'D'
        else: grade = 'F'
            
        answer += grade
    return answer
```

```python
Q1 = [[100,90,98,88,65],[50,45,99,85,77],[47,88,95,80,67],[61,57,100,80,65],[24,90,94,75,65]]
print('Q1.')
print(f'Answer : {solution(Q1)}')
```
