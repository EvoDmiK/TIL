```python
ROOT_PATH = '/'.join(os.getcwd().split('/')[:-1])
DATA_PATH = f'{ROOT_PATH}/data'
```

```python
def jaccad_sim(word1, word2):
    
    word1, word2 = set(word1), set(word2)
    inter = word1.intersection(word2)
    union = word1.union(word2)
    
    try: return len(inter) / len(union)
    except: return 0.0
```

```python
import numpy as np

words = open(f'{DATA_PATH}/counter.txt', 'r', encoding='UTF-8').readlines()
dictionary = open(f'{DATA_PATH}/thyroid_dict.txt', 'r').readlines()

texts = ''
results = []
result_append = results.append
for odx, word in enumerate(words, 1):
    # print(f'[{odx}] {word}')
    
    word = word.replace('\n', '')
    scores, names = [], []
    score_append = scores.append
    name_append = names.append
    
    for idx, name in enumerate(dictionary, 1):
        name = name.replace('\n', '')
        score = jaccad_sim(word, name)
        
        condition = [word == '', score < 0.5]
        if any(condition): continue
        
        score_append(score); name_append(name)

    try:
        score_idx = np.argmax(scores)
        max_score = scores[score_idx]
        max_name  = names[score_idx]
        
        text = f'[{odx}-{idx}]\nword1 : {word} word2 : {max_name} ({max_score*100:.2f}%) \n'
        # print(text)
        
        result_append((max_score, f'{max_name}:{word}'))

    except:
        pass

    
```

```python
result_dict = sorted({result[1] : result[0] for result in results}.items(), key = lambda x: x[1], reverse = True)
result_dict = {result[0] : result[1] for result in result_dict}

for odx, (name, score) in enumerate(result_dict.items(), 1):
    
    try:
        if score * 100 >= 80.0:
            name, word = name.split(':')
            print(f'[{str(odx).zfill(4)}] word : {word}, name : {name} ({score*100:.2f}%)')
            scores.append(score*100)

    except Exception as e:
        print(name)
        print(e)
```
