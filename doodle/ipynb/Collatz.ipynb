```python
import matplotlib.pyplot as plt
from time import time
```

```python
def collatz(n):
    
    cnt, nums = 0, [n]
    while n > 1:
        n = n // 2 if n % 2 == 0 else 3*n + 1
        cnt += 1
        nums.append(n)
        
    return nums, cnt
```

```python
exp_dict     = {'n' : [], 'nums' : [], 'cnts' : []}
elapsed_time = 0
iters = 10**8

for idx in range(1, iters):
    
    start_time = time()
    nums, cnt  = collatz(idx)
    
    exp_dict['n'].append(idx)
    exp_dict['nums'].append(nums)
    exp_dict['cnts'].append(cnt)
    
    elapsed_time += time() - start_time
    
    if idx % 10**5 == 0: 
        print(f'[INFO] [{idx:,} / {iters:,}] ({elapsed_time:.2f} sec per 10^5 iter) [{idx / iters:.3f}%]')
        elapsed_time = 0
    
```

```python
index = 999999
plt.figure(figsize = (10, 10))
plt.plot(range(len(exp_dict['nums'][index])), exp_dict['nums'][index])
plt.xticks([idx * 10 for idx in range(len(exp_dict['nums'][index]) // 10 + 1)])
```

```python
exp_dict['nums'][index][-5:] == [16, 8, 4, 2, 1]
```
