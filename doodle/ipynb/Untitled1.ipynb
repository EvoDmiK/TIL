```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def circle(a, b, r):
    
    x = [a + r*np.cos(theta) for theta in np.arange(0, 2*np.pi, 0.01)]
    y = [b + r*np.sin(theta) for theta in np.arange(0, 2*np.pi, 0.01)]
    
    return (x, y)


line = lambda x: 13*x  - 67
```

```python
x1, y1 = circle(1, 1, 13)
x2, y2 = circle(40, -2, 37)

x_l    = np.arange(2, 8, 0.01)
y_l    = line(x_l)
```

```python
plt.plot(x1, y1, x2, y2, x_l, y_l)
```
