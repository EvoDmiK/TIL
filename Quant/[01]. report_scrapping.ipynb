```python
from typing import List

from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
url      = "http://www.paxnet.co.kr/stock/report/report?wlog_rpt=jm&menuCode=2222"
response = requests.get(url)
html     = response.content
html     = bs(html)
```

```python
div      = html.find('div', attrs = {'class' : 'board-type'})
li_list  = div.find_all('li')

li_list[0]
```

```python
def parsing_li(li:str) -> List:
    ## 종목 이름
    stock = li.find_all('a')[0].text
    
    ## 제목
    title = li.find_all("a")[1].text.replace('\y', '')

    meta_data = li.find_all('div', attrs = {'class' : 'line3'})
                        
    ## 적정 가격
    price   = meta_data[0].text
    price   = [p for p in price.split('\n') if '원' in p][0].replace('\t', '').replace('\r', '')
    
    ## 의견
    opinion = meta_data[1].text
    opinion = opinion.split('\n')[1].replace('\t', '').replace('\r', '')

    ## 증권사
    trading_firm = meta_data[2].text
    
    ## 날짜
    date        = meta_data[3].text
    return [stock, title, price, opinion, trading_firm, pd.to_datetime(date)]
```

```python
datas = []
for li in li_list:
    
    ## 종목 이름인지 아닌지 확인하는 코드
    check_str = '<strong class="color-cate"><span class="bracket">'
    if check_str in str(li):
        datas.append(parsing_li(li))

columns = ['종목명', '리포트제목', '적정가격', '의견', '증권사', '날짜']
df      = pd.DataFrame(datas, columns = columns)

df
```
