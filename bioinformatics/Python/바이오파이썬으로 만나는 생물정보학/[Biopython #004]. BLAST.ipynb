# 1. BLAST (Basic Local Alignment Search Tool)
- DNA 염기서열 또는 단백질의 아미노산 서열을 서로 비교하기 위한 알고리즘  
- 대용량의 서열과 입력서열을 비교하여 가장 유사한 서열을 검색할 수 있다.  
- 뉴클레오티드-뉴클레오티드, 단백질-뉴클레오티드, 뉴클레오티드-단백질,  
  단백질-단백질 등의 쿼리, DB 쌍에 대한 버전을 가지고 있는 프로그램의 집합  
- 염기로 구성된 서열이 입력일 경우, blastn, tblastx, blatx를 이용  
- 단백질로 구성된 서열이 입력일 경우 blastp, tblastn을 이용  

|프로그램|설명|
|:---:|:---:|
|blastn| Nucleotide sequence 간 비교|
|tblastx| Query nucleotide sequence와 nucleotide sequence DB를 6-frame으로 변환 후 비교|
|blastx| Query nucleotide sequence를 6-frame으로 변환 후 protein sequnece db와 비교|
|blastp|Protein sequence 간 비교|
|tblastn|Nucleotide sequence DB를 6-frame으로 변환 후 query protein sequence와 비교|

**(!) 참조한 논문에 의하면 유전자 서열을  번역하여 단백질 서열을 비교하는 것이 더 정확하다고 하는데, 다음과 같은 이유들 때문이다.**
   1) 유전자 서열은 4 종류로 구성되어 있지만, 단백질 서열은 20개의 아미노산 서열로 구성되어 있다.  
      i.e. 우연히 두 residue가 일치한 확률이 단백질이 1/20인 반면에 유전자의 경우는 1/4이다.  
      
   2) 유전자의 coding 특성에 의해 다른 유전자가 같은 아미노산을 coding 할 수 있다.  
      i.e. 단백질 level에서 같은 서열을 가짐에도 불구하고 유전자 서열에서는 많은 차이가 날 수 있다.

## 1-2. 바이오파이썬에서의 BLAST 실행
- 바이오파이썬에서 BLAST를 실행하기 위해서는 NCBIWWW 모듈을 사용한다.  
- NCBIWWW 모듈의 qblast() 메소드를 사용하며, 3가지 인자가 들어간다.  
  1) BLAST 프로그램명 : blastn, blastp, blastx, tblastn, tblastx를 지원
  2) BLAST 데이터 베이스 : nt, nr
  3) BLAST를 수행할 서열 : 서열 자체나 FASTA 파일이 들어갈 수 있다.
- NCBIWWW.qblast()를 실행하여 BLAST를 진행하는 과정은 크게 2가지로 나뉜다.
  1) 서열을 넣어 BLAST 실행
  2) BLAST 실행하여 나온 결과 파싱
  
**추가적으로 biopython 공식 사이트에서는 qblast 메소드의 유용한 인자값을 몇개 더 소개하였다.**
1) ncbi_gi : 'gi' identifier를 줄 것인지 아닌지 (True / False)
2) descriptions : BLAST 결과에서 보여지는 description 갯수 (기본값 500)
3) alignments : BLAST 결과에서 보여지는 alignment 갯수 (기본값 500)
4) expect : E-Value 임계값 (기본값 10.0)
5) matrix_name : Scoring matrix (PAM30, PAM70, BLOSUM80, BLOSUM45)
6) filter : 필터링 끄기 (기본값 None, 꺼져있음)
7) format_type : BLAST 결과가 반환될 형식 (HTML, Text, ASN.1, XML | 기본값 XML)
8) entrez_query : Entrez query to limit Blast search
9) hitlist_size : Number of hits to return. (기본값 50)
10) megablast : whether to use MEga BLAST algorithm (blastn only) (True / False)
11) service : plain, psi, phi, rpsblast, megablast 중 소문자로 입력
### 1-2-1. NCBIWWW.qblast() 실행
- 이 과정은 반드시 인터넷에 연결되어 있어야 한다.
- 실행된 결과를 readlines로 읽으면 XML 형식의 <태그> 형태의 문자열 결과물이 나온다.  

```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO

record = SeqIO.read("../../../Dataset/biopython/fasta/buccal_swab.unmapped1.fasta", format = 'fasta')

## 이 부분에서 시간이 좀 걸린다..
handle = NCBIWWW.qblast('blastn', 'nt', record.format('fasta'))

results = handle.readlines()[:30]
for result in results:
    print(result)
```

### 1-2-2. BLAST 결과 XML 파일 파싱하기
- BLAST 실행결과로 나온 XML 파일을 파싱하는 방법은  
  NCBIXML.parse()메소드를 이용한다.
- NCBIXSML.parse() 메소드 안에 인자값으로 BLAST를 수행한 결과를 입력해준다.

### (!) BLAST 결과 분석
- DB에 매치된 결과가 특별한 의미를 갖고 있는지 알아보기 위해 주로 E-value를 사용한다.    
- E-value는 특정 사이즈의 DB 안에서 매칭을 수행할 때 우연히 하나의 매치가 발생할 기대값으로 정의됨.  
- 우연히 매치되는 확률이 적을 때 (i.e. E-value 값이 작을때) 더 의미있다고 할 수 있음.  
- 매치의 퀄리티는 E-value 뿐 아니라, alignment의 길이와 percentage similarity에도 좌우된다.

**아래 코드를 실행시킨 결과는 웹 페이지에서 BLAST를 한 것과 동일하게 Streptococcus pneumoniae strain이 나왔다.**

```python
from Bio.Blast import NCBIXML

handle = NCBIWWW.qblast('blastn', 'nt', record.format('fasta'))
blast_records = NCBIXML.parse(handle)

E_VALUE_THRESHOLD = 0.05

## 검색 결과가 굉장히 많다보니 출력 결과도 굉장히 길다.. 
for blast_record in blast_records:
    for idx, alignment in enumerate(blast_record.alignments[:10], 1):
        for hsp in alignment.hsps:
            if hsp.expect < E_VALUE_THRESHOLD:
                print(f'[{idx}]')
                print(f'alignment 제목 : {alignment.title}')
                print(f'alignment 길이 : {alignment.length}')
                print(f'E-Value        : {hsp.expect}')
                print(f'쿼리 염기서열   : {hsp.query[0:75]}')
                print(f'매칭           : {hsp.match[0:75]}')
                print(f'검색 결과      : {hsp.sbjct[0:75]}')
                print('\n')
```
