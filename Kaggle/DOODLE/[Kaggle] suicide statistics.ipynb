```python
import os

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
SEP       = os.path.sep
ROOT_PATH = SEP.join(os.getcwd().split(SEP)[:-2])
DATA_PATH = f'{ROOT_PATH}/Dataset/suicide/who_suicide_statistics.csv'
```

```python
df = pd.read_csv(DATA_PATH)
df.rename({'suicides_no' : 'suicides'}, inplace = True, axis = 1)

df['suicides'].fillna(0, inplace = True)
df.dropna(inplace = True)
```

```python
df[df['sex'] == 'female']['year'].value_counts()
```

```python
df_year   = df.groupby('year').sum().sort_values(by = 'year', ascending = False)
df_year
```

```python
anomaly_data = df[df.year == 2016]
anomaly_data
```

```python
df.drop(anomaly_data.index, axis = 0, inplace = True)
df_year = df.groupby('year').sum().sort_values(by = 'year')
df_year
```

```python
plt.figure(figsize = (16, 8))
plt.bar(np.arange(len(df_year)), df_year.suicides)
plt.xticks(np.arange(len(df_year)), (df_year.index), rotation = 45)
plt.show()
```
