```python
from collections import Counter
import ast
import os

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ROOT_PATH = '/'.join(os.getcwd().split('/')[:-3])
DATA_PATH = f'{ROOT_PATH}/AI_study/dataset/Netflix-Family/netflix-2022'
```

```python
def counting(datas):
    
    total_data = ''
    for data in datas: total_data += f' {data}'
    
    return Counter(total_data.split(' '))        
```

```python
df_credit = pd.read_csv(f'{DATA_PATH}/raw_credits.csv')
df_title  = pd.read_csv(f'{DATA_PATH}/raw_titles.csv')

df_merge  = pd.merge(df_credit, df_title, how='left', on='id')
df_merge.info()
```

```python
df_merge  = df_merge[['person_id', 'id', 'name', 'title', 'type', 'production_countries', 'genres', 'release_year']]
df_merge.head()
```

```python
df_merge['production_countries'] = df_merge['production_countries'].apply(lambda x: ' '.join(ast.literal_eval(x)))
df_merge['genres']               = df_merge['genres'].apply(lambda x: ' '.join(ast.literal_eval(x)))
```

```python
df_koreans = df_merge[df_merge['production_countries'] == 'KR']
```

```python
total_genres = ''
for data in df_koreans['genres']: total_genres += f' {data}'
total_genres = [genre for genre in total_genres.split(' ') if genre != '']

sorted(Counter(total_genres).items(), key = lambda x: x[1], reverse = True)
```
