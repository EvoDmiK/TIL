```python
def max_path(triangle):
    sum_path = 0
    
    for idx, nums in enumerate(triangle):

        if idx == 0: max_idx = nums.index(max(nums))
        
        
        search_seq = nums[:max_idx + 2] if max_idx == 0 else \
                     nums[max_idx : max_idx + 2]
        
        max_idx    = search_seq.index(max(search_seq))
        max_val    = max(search_seq)

        sum_path += max_val

    return sum_path
```

```python
def fixed_max_path(triangle):
    
    n    = len(triangle)
    
    ## 삼각형의 바닥부터 올라가면서 더해줌.
    for i, leaves in enumerate(triangle[-2::-1]):
        
        ## 왼쪽 데이터랑 오른쪽 데이터랑 두 개씩 잘라 비교하여 
        ## 큰 값을 탐색하는 줄의 윗 줄에 더해줌.
        for j, _ in enumerate(leaves):
            
            left  = triangle[n - i - 1][j]
            right = triangle[n - i - 1][j + 1]
            
            print(triangle, left,right)
            
            triangle[n - i - 2][j] += max(left, right)
            
            print(triangle, left,right, '\n')
    return triangle[0][0]
```

```python
triangle  = [
                  [3], 
                 [7, 4], 
                [2, 4, 6], 
               [8, 5, 9, 3]
            ]

print(f'Dove\'s Method  : {max_path(triangle)}')
print(f'Fixed Method   : {fixed_max_path(triangle)}')
```
