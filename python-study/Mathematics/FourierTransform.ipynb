```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
domain = np.arange(-2*np.pi, 2*np.pi, 0.001)
eps    = 1e-6

sin = np.sin(domain)
cos = np.cos(domain)
```

```python
plt.axvline(x=0, color = 'black')  # draw x =0 axes
plt.axhline(y=0, color = 'black')   # draw y =0 axes

plt.plot(domain, sin)
plt.plot(domain, cos)
```

```python
plt.axvline(x=0, color = 'black')  # draw x =0 axes
plt.axhline(y=0, color = 'black')   # draw y =0 axes

func1 = lambda x: 2*x
func2 = lambda x: 2*(1 - x)
func3 = lambda x: 2*(x - 2)

y1 = func1(np.arange(0, 0.5, eps))
y2 = func2(np.arange(0.5, 1.5, eps))
y3 = func3(np.arange(1.5, 2, eps))

x = np.arange(0, 2, eps)
y = np.concatenate([y1, y2, y3])

plt.plot(x, y)
```

```python
cycle, coef  = 2, 0.5
omega = 2*np.pi / cycle

fourier_y = []
for idx in range(cycle): fourier_y.append(coef*np.exp(1j*omega*idx**2))
```
